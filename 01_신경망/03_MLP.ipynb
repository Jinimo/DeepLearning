```python
import numpy as np
```

1. `And` gate


```python
weight = np.array([1,1]) # 가중치
bias = -1                # 편향 
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  sum = weight[0] * x1 + weight[0] * x2 + bias # 가중 합 
  print(0 if sum <= 0 else 1)                  # 활성화 함수
```

```python
# dot product 활용 
weight = np.array([1,1]) # 가중치
bias = -1                # 편향 
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  x = np.array([x1, x2])
  # 가중 합
  sum = np.dot(weight, x) + bias  # 가중 합 
  print(0 if sum <= 0 else 1)     # 활성화 함수
```

```python
# 함수 만들기
def AND_gate(x1, x2):
  weight = np.array([1,1]) # 가중치
  bias = -1                # 편향 
  x = np.array([x1, x2])
  sum = np.dot(weight, x) + bias  # 가중 합
  return 0 if sum <= 0 else 1     # 활성화 함수
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  print(AND_gate(x1, x2))
```

2. `OR` gate

```python
# 함수 만들기
def OR_gate(x1, x2):
  weight = np.array([2,2]) # 가중치
  bias = -1                # 편향 
  x = np.array([x1, x2])
  sum = np.dot(weight, x) + bias  # 가중 합
  return 0 if sum <= 0 else 1     # 활성화 함수
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  print(OR_gate(x1, x2))
```

3. `NAND` gate


```python
def NAND_gate(x1, x2):
  weight = np.array([-2,-2]) # 가중치
  bias = 3               # 편향 
  x = np.array([x1, x2])
  sum = np.dot(weight, x) + bias  # 가중 합
  return 0 if sum <= 0 else 1     # 활성화 함수
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  print(NAND_gate(x1, x2))
```

# MLP

```python
n1, n2 = [], []
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  n1.append(NAND_gate(x1, x2))
  n2.append(OR_gate(x1, x2))
```

```python
for x1, x2 in zip(n1, n2):
  print(AND_gate(x1, x2))
```

### `XOR gate`
- XOR 문제 해결 위해 Weight 6개, bias 3개 필요
- 이를 만족하는 Weight, bias 조합 무수히 많음 


```python
# XOR_gate 함수 만들기
# 1-layer : NAND, OR gate
# 2-layer : AND gate
def XOR_gate(x1, x2):
  return AND_gate(NAND_gate(x1, x2), OR_gate(x1, x2))
```
