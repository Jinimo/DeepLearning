# Perceptron
- 입력 값과 활성화 함수를 사용해 출력값을 자음으로 넘기는 가장 작은 신경망 단위


```python
import numpy as np
```

1. `And` gate


```python
weight = np.array([1,1]) # 가중치
bias = -1                # 편향 
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  sum = weight[0] * x1 + weight[0] * x2 + bias # 가중 합 
  print(0 if sum <= 0 else 1)                  # 활성화 함수
```

```python
# dot product 활용 
weight = np.array([1,1]) # 가중치
bias = -1                # 편향 
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  x = np.array([x1, x2])
  # 가중 합
  sum = np.dot(weight, x) + bias  # 가중 합 
  print(0 if sum <= 0 else 1)     # 활성화 함수
```

```python
# 함수 만들기
def And_gate(x1, x2):
  weight = np.array([1,1]) # 가중치
  bias = -1                # 편향 
  x = np.array([x1, x2])
  sum = np.dot(weight, x) + bias  # 가중 합
  return 0 if sum <= 0 else 1     # 활성화 함수
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  print(And_gate(x1, x2))
```

2. `OR` gate

```python
# 함수 만들기
def OR_gate(x1, x2):
  weight = np.array([2,2]) # 가중치
  bias = -1                # 편향 
  x = np.array([x1, x2])
  sum = np.dot(weight, x) + bias  # 가중 합
  return 0 if sum <= 0 else 1     # 활성화 함수
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  print(OR_gate(x1, x2))
```

3. `NAND` gate


```python
def NAND_gate(x1, x2):
  weight = np.array([-2,-2]) # 가중치
  bias = 3               # 편향 
  x = np.array([x1, x2])
  sum = np.dot(weight, x) + bias  # 가중 합
  return 0 if sum <= 0 else 1     # 활성화 함수
```

```python
for x1, x2 in [(0,0), (0,1), (1,0), (1,1)]: 
  print(NAND_gate(x1, x2))
```
